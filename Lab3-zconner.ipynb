## Import necessary python packages

```python
# Import the necessary python packages and use aliases wwhen necessary
import os

import matplotlib.pyplot as plt

# use print functiont to show a message that import was sucessful
print("import was sucessful")
```

## Creating Variables 

```python
#Create variables for monthly average temperature (Fahrenheit) for January 
#through December
Jan = 38.0
Feb = 41.6
Mar = 39.2
Apr = 57.2
May = 61.1
Jun = 72.0
Jul = 76.8
Aug = 74.0
Sep = 70.5
Oct = 64.1
Nov = 46.6
Dec = 33.4

##Print the variable for December
print(Dec)
```

## Run calculations on variables

```python
#Convert the variables from Fahrenheit to Celsius using 
#Celsius = (Farhenheit - 32)/1.8
#NOTE we will assign these converted values to the original variable names ##
Jan = (38.0 - 32.0)/1.8
Feb = (41.6 - 32.0)/1.8
Mar = (39.2 - 32.0)/1.8
Apr = (57.2 - 32.0)/1.8
May = (61.1 - 32.0)/1.8
Jun = (72.0 - 32.0)/1.8
Jul = (76.8 - 32.0)/1.8
Aug = (74.0 - 32.0)/1.8
Sep = (70.5 - 32.0)/1.8
Oct = (64.1 - 32.0)/1.8
Nov = (46.6 - 32.0)/1.8
Dec = (33.4 - 32.0)/1.8

#Print the newly converted value for December
print(Dec)
```

## Create lists

```python
#Create a list containing the newly converted celsius values
temp = [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]

#Print the temperature list
print(temp)

#Create a list of the months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

#Print the months list
print(months)
```

## Plot data from lists

```python
#Create a green line plot of the average monthly temp for Central Park, NY using the lists above

#set the plot size for all plots that follow
plt.rcParams["figure.figsize"] = (8, 8)

#set plot title size for all plots that follow
plt.rcParams['axes.titlesize'] = 20

#create the plot space upon which to plot the data
fig, ax = plt.subplots()

#add the x-axis and the y-axis to the plot
ax.plot(months, temp, color="green")

#set plot title
ax.set(title="2017 Average Monthly Temperature in Central Park, New York")

#add labels to the axes
ax.set(xlabel="Month", ylabel="Temperature (Celsius)");
```

## Plot data from lists II

```python
#Create a red bar plot of the average monthly temperature for Central Park, NY using the lists above

#set plot size for all plots that follow
plt.rcParams["figure.figsize"] = (8, 8)

#create the plot space upon which to plot the data
fig, ax = plt.subplots()

#add the x-axis and the y-axis to the plot
ax.bar(months, temp, color="red")

#set plot title
ax.set(title="2017 Averge Monthly Temperature in Central Park, New York")

#add labels to the axes
ax.set(xlabel="Month", ylabel= "Temperature (Celsius)");
```

## Plot data from lists III

```python
#Create a blue scatterplot of the average monthly temperature for Central Park, NY using the lists above

#set plot size for all plots that follow
plt.rcParams["figure.figsize"] = (8, 8)

#create the plot space upon which to plot the data
fig, ax = plt.subplots()

#plot a scatter plot
ax.scatter(months, temp, color="blue")

#set plot title
ax.set(title="2017 Average Monthly Temperature in Central Park, New York")

#add labels to the axes
ax.set(xlabel="Month", ylabel="Temperature (Celsius)");
```
